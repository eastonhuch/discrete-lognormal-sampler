```python
from generate_data import generate_data
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = generate_data()
```

```python
# Show mean and variance functions together
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 3))

ax1.plot(data.days, data.mu, label="mean")
ax1.set_xlabel("Date")
ax1.set_ylabel("Event Count")
ax1.set_title("E(Z)")

ax2.plot(data.days, data.sigma, label="standard deviation", color="green")
ax2.set_xlabel("Date")
plt.ylabel("Event Count")
ax2.set_title("sd(Z)")

plt.savefig("./figures/mean_var_functions.png")
plt.show()
```

```python
# Plot y
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(2, 1, 1)
ax.plot(data.days, data.y)
plt.ylim((0, 1.1*np.max(data.y)))
plt.xlabel("Date")
plt.ylabel("Event Count")
plt.title("Events Over Time")
plt.savefig("./figures/events_over_time.png")
plt.show()
```
